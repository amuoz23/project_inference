```python
import pandas as pd

# Cargar los datos en un DataFrame
# Aquí reemplaza 'base_datos.csv' con el nombre de tu archivo CSV
df = pd.read_csv('data_math4.csv')

# Función para ajustar los valores de la columna PROMEDIOCARRERA al rango 0-5
def ajustar_promedio(promedio):
    # Si el valor está entre 0 y 5, lo dejamos tal cual
    if 0 <= promedio <= 5:
        return round(promedio, 2)
    # Si el valor es mayor a 1000, lo dividimos entre 1000 para ajustarlo
    elif promedio >= 1000:
        return round(promedio / 1000, 3)
    # Si el valor está entre 100 y 1000, lo dividimos entre 100
    elif promedio >= 100:
        return round(promedio / 100, 3)
    # Si el valor es mayor a 10, lo dividimos entre 10
    elif promedio >= 10:
        return round(promedio / 10, 3)
    return promedio

# Aplicar la función a la columna 'PROMEDIOCARRERA'
df['PROMEDIOSEMESTRE'] = df['PROMEDIOSEMESTRE'].apply(ajustar_promedio)

# Guardar la base de datos corregida en un nuevo archivo CSV
# Aquí reemplaza 'base_datos_corregida.csv' con el nombre que deseas para el nuevo archivo
df.to_csv('base_datos_corregida.csv', index=False)

# Mostrar el DataFrame resultante
print(df)
```

```python
import pandas as pd

# Cargar los datos en un DataFrame
# Reemplaza 'base_datos.csv' con el nombre de tu archivo CSV
df = pd.read_csv('data_math5.csv')

# Función para verificar si un valor está dentro del rango de 0 a 5
def verificar_rango(promedio):
    if 0 <= promedio <= 5:
        return True
    else:
        return False

# Recorremos la columna 'PROMEDIOCARRERA' y verificamos si los valores están en el rango
out_of_range = []
for idx, value in enumerate(df['PROMEDIOSEMESTRE']):
    if not verificar_rango(value):
        out_of_range.append(idx)

# Si encontramos valores fuera de rango, los mostramos
if out_of_range:
    print(f"Los siguientes índices tienen valores fuera del rango 0-5:")
    for idx in out_of_range:
        print(f"Índice: {idx}, Valor: {df['PROMEDIOSEMESTRE'][idx]}")
else:
    print("Todos los valores están dentro del rango 0-5.")
```

```python
import pandas as pd

# Cargar los datos en un DataFrame
# Reemplaza 'base_datos.csv' con el nombre de tu archivo CSV
df = pd.read_csv('data_math5.csv')

# Función para etiquetar el estado actual del estudiante
def etiquetar_estado(estado):
    estado = str(estado).strip().upper()  # Convertir a cadena, eliminar espacios extra y convertir a mayúsculas
    if 'ACTIVO' in estado:
        return 0  # Estudiante activo
    elif 'INACTIVO' in estado or 'INACTIVO MATRICULA FINANCIERA' in estado:
        return 1  # Estudiante inactivo, incluyendo los casos "INACTIVO DOS PROGRAMAS", "INACTIVO MATRICULA FINANCIERA"
    elif 'GRADUADO' in estado:
        return 0  # Estudiante graduado
    else:
        return None  # En caso de que haya un estado desconocido

# Crear la nueva columna 'ETIQUETA_ESTADO' con las etiquetas correspondientes
df['ETIQUETA_ESTADO'] = df['ESTADO_ACTUAL_ESTUDIANTE'].apply(etiquetar_estado)

# Guardar el DataFrame modificado en un nuevo archivo CSV (opcional)
# Reemplaza 'base_datos_etiquetada.csv' con el nombre del archivo de salida
df.to_csv('base_datos_etiquetada.csv', index=False)

# Mostrar las primeras filas del DataFrame con la nueva columna 'ETIQUETA_ESTADO'
print(df[['ESTADO_ACTUAL_ESTUDIANTE', 'ETIQUETA_ESTADO']].head())
```

```python
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Cargar datos (asegúrate de usar el nombre real de tu archivo)
df = pd.read_csv("data_fin_KM.csv")

# Variables clave
durations = df["TIEMPO"]  # Tiempo de observación
event_observed = df["ETIQUETA"]  # 1 si desertó, 0 si no

# Crear el objeto Kaplan-Meier
kmf = KaplanMeierFitter()

# Ajustar el modelo
kmf.fit(durations, event_observed, label="Supervivencia general")

# Graficar la curva de supervivencia
kmf.plot_survival_function()
plt.title("Curva de Supervivencia - Kaplan-Meier")
plt.xlabel("Tiempo (semestres)")
plt.ylabel("Probabilidad de supervivencia")
plt.grid(True)
plt.show()
```
