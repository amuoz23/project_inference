```python
import pandas as pd

# Cargar los datos en un DataFrame
# Aquí reemplaza 'base_datos.csv' con el nombre de tu archivo CSV
df = pd.read_csv('data_math4.csv')

# Función para ajustar los valores de la columna PROMEDIOCARRERA al rango 0-5
def ajustar_promedio(promedio):
    # Si el valor está entre 0 y 5, lo dejamos tal cual
    if 0 <= promedio <= 5:
        return round(promedio, 2)
    # Si el valor es mayor a 1000, lo dividimos entre 1000 para ajustarlo
    elif promedio >= 1000:
        return round(promedio / 1000, 3)
    # Si el valor está entre 100 y 1000, lo dividimos entre 100
    elif promedio >= 100:
        return round(promedio / 100, 3)
    # Si el valor es mayor a 10, lo dividimos entre 10
    elif promedio >= 10:
        return round(promedio / 10, 3)
    return promedio

# Aplicar la función a la columna 'PROMEDIOCARRERA'
df['PROMEDIOSEMESTRE'] = df['PROMEDIOSEMESTRE'].apply(ajustar_promedio)

# Guardar la base de datos corregida en un nuevo archivo CSV
# Aquí reemplaza 'base_datos_corregida.csv' con el nombre que deseas para el nuevo archivo
df.to_csv('base_datos_corregida.csv', index=False)

# Mostrar el DataFrame resultante
print(df)
```

```python
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Cargar datos (asegúrate de usar el nombre real de tu archivo)
df = pd.read_csv("data_fin_KM.csv")

# Variables clave
durations = df["TIEMPO"]  # Tiempo de observación
event_observed = df["ETIQUETA"]  # 1 si desertó, 0 si no

# Crear el objeto Kaplan-Meier
kmf = KaplanMeierFitter()

# Ajustar el modelo
kmf.fit(durations, event_observed, label="Supervivencia general")

# Graficar la curva de supervivencia
kmf.plot_survival_function()
plt.title("Curva de Supervivencia - Kaplan-Meier")
plt.xlabel("Tiempo (años)")
plt.ylabel("Probabilidad de supervivencia")
plt.grid(True)
plt.show()
```

```python
# Dividir por género
male = df[df["GENERO"] == 1]
female = df[df["GENERO"] == 0]

# Kaplan-Meier para hombres
kmf_male = KaplanMeierFitter()
kmf_male.fit(male["TIEMPO"], male["ETIQUETA"], label="Hombres")
kmf_male.plot_survival_function()

# Kaplan-Meier para mujeres
kmf_female = KaplanMeierFitter()
kmf_female.fit(female["TIEMPO"], female["ETIQUETA"], label="Mujeres")
kmf_female.plot_survival_function()

# Personalizar gráfico
plt.title("Curva de Supervivencia por Género")
plt.xlabel("Tiempo (años)")
plt.ylabel("Probabilidad de supervivencia")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from lifelines import KaplanMeierFitter, statistics
import matplotlib.pyplot as plt

# Función para comparar supervivencias entre grupos
def comparar_supervivencia(df, variable, etiquetas_grupos, columna_tiempo="TIEMPO", columna_evento="ETIQUETA"):
    plt.figure(figsize=(10, 6))
    kmf = KaplanMeierFitter()
    
    for valor, etiqueta in zip(df[variable].unique(), etiquetas_grupos):
        grupo = df[df[variable] == valor]
        kmf.fit(grupo[columna_tiempo], grupo[columna_evento], label=f"{etiqueta}")
        kmf.plot_survival_function()

    plt.title(f"Curvas de Supervivencia según {variable}")
    plt.xlabel("Tiempo (años)")
    plt.ylabel("Probabilidad de supervivencia")
    plt.legend()
    plt.grid(True)
    plt.show()

# Comparación 1: Por género
comparar_supervivencia(df, variable="GENERO", etiquetas_grupos=["Femenino", "Masculino"])

# Comparación 2: Por estrato socioeconómico
comparar_supervivencia(df, variable="ESTRATO", etiquetas_grupos=[f"Estrato {e}" for e in df["ESTRATO"].unique()])

# Comparación 3: Promedio de carrera (Dividir en alto/bajo)
df["PROMEDIO_ALTO_BAJO"] = (df["PROMEDIOCARRERA"] >= df["PROMEDIOCARRERA"].median()).astype(int)
comparar_supervivencia(df, variable="PROMEDIO_ALTO_BAJO", etiquetas_grupos=["Promedio Bajo", "Promedio Alto"])

# Comparación 4: Periodo de ingreso (antes/después de 2010)
df["COHORTE"] = (df["PERIODO_INGRESO"] >= 2016).astype(int)
comparar_supervivencia(df, variable="COHORTE", etiquetas_grupos=["Antes de 2016", "2016 o después"])

# Test de log-rank para diferencias significativas
from lifelines.statistics import logrank_test

# Ejemplo: Género
male = df[df["GENERO"] == 1]
female = df[df["GENERO"] == 0]
result_gender = logrank_test(male["TIEMPO"], female["TIEMPO"], event_observed_A=male["ETIQUETA"], event_observed_B=female["ETIQUETA"])
print(f"Diferencias por género (p-valor): {result_gender.p_value}")

# Ejemplo: Estrato socioeconómico
estrato_1 = df[df["ESTRATO"] == 1]
estrato_2 = df[df["ESTRATO"] == 2]
result_estrato = logrank_test(estrato_1["TIEMPO"], estrato_2["TIEMPO"], event_observed_A=estrato_1["ETIQUETA"], event_observed_B=estrato_2["ETIQUETA"])
print(f"Diferencias por estrato (p-valor): {result_estrato.p_value}")
```

```python
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
import matplotlib.pyplot as plt

# Función para comparar supervivencias entre grupos y calcular p-valores
def comparar_supervivencia_y_pvalores(df, variable, etiquetas_grupos, columna_tiempo="TIEMPO", columna_evento="ETIQUETA"):
    plt.figure(figsize=(10, 6))
    kmf = KaplanMeierFitter()
    grupos = []
    
    for valor, etiqueta in zip(df[variable].unique(), etiquetas_grupos):
        grupo = df[df[variable] == valor]
        grupos.append(grupo)
        kmf.fit(grupo[columna_tiempo], grupo[columna_evento], label=f"{etiqueta}")
        kmf.plot_survival_function()

    # Comparar todas las combinaciones de grupos con el test log-rank
    p_values = []
    for i in range(len(grupos)):
        for j in range(i + 1, len(grupos)):
            result = logrank_test(
                grupos[i][columna_tiempo], grupos[j][columna_tiempo],
                event_observed_A=grupos[i][columna_evento],
                event_observed_B=grupos[j][columna_evento]
            )
            p_values.append((etiquetas_grupos[i], etiquetas_grupos[j], result.p_value))
    
    # Mostrar los p-valores
    print(f"P-valores para la comparación según {variable}:")
    for g1, g2, p in p_values:
        print(f"  {g1} vs {g2}: p = {p:.4f}")
    
    plt.title(f"Curvas de Supervivencia según {variable}")
    plt.xlabel("Tiempo (años)")
    plt.ylabel("Probabilidad de supervivencia")
    plt.legend()
    plt.grid(True)
    plt.show()

# Comparación 1: Por género
comparar_supervivencia_y_pvalores(
    df, variable="GENERO", etiquetas_grupos=["Femenino", "Masculino"]
)

# Comparación 2: Por estrato socioeconómico
comparar_supervivencia_y_pvalores(
    df, variable="ESTRATO", etiquetas_grupos=[f"Estrato {e}" for e in df["ESTRATO"].unique()]
)

# Comparación 3: Promedio de carrera (Dividir en alto/bajo)
df["PROMEDIO CARRERA"] = (df["PROMEDIOCARRERA"] >= df["PROMEDIOCARRERA"].median()).astype(int)
comparar_supervivencia_y_pvalores(
    df, variable="PROMEDIO CARRERA", etiquetas_grupos=["Promedio Bajo", "Promedio Alto"]
)
# Comparación 3: Promedio de semestre (Dividir en alto/bajo)
df["PROMEDIO SEMESTRE"] = (df["PROMEDIOSEMESTRE"] >= df["PROMEDIOSEMESTRE"].median()).astype(int)
comparar_supervivencia_y_pvalores(
    df, variable="PROMEDIO SEMESTRE", etiquetas_grupos=["Promedio Bajo", "Promedio Alto"]
)

# Comparación 4: Periodo de ingreso (antes/después de 2010)
df["COHORTE"] = (df["PERIODO_INGRESO"] >= 2010).astype(int)
comparar_supervivencia_y_pvalores(
    df, variable="COHORTE", etiquetas_grupos=["Antes de 2010", "2010 o después"]
)
```

```python
from lifelines import CoxPHFitter

# Selección de columnas relevantes
columnas_modelo = [
    "TIEMPO", "ETIQUETA", "GENERO", "ESTRATO", "SEMESTRE_ACTUAL",
    "PROMEDIOCARRERA", "PROMEDIOSEMESTRE", "COHORTE"
]
df_modelo = df[columnas_modelo]

# Ajustar el modelo de Cox
cph = CoxPHFitter()
cph.fit(df_modelo, duration_col="TIEMPO", event_col="ETIQUETA")
cph.print_summary()  # Mostrar resultados del modelo

# Visualización de coeficientes y análisis
cph.plot()  # Visualizar coeficientes de riesgos proporcionales
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Selección de variables predictoras y objetivo
X = df[["GENERO", "ESTRATO", "PROMEDIOCARRERA", "PROMEDIOSEMESTRE", "COHORTE"]]
y = df["ETIQUETA"]

# Dividir en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Ajustar modelo de regresión logística
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Evaluación del modelo
y_pred = logreg.predict(X_test)
print("Reporte de Clasificación:")
print(classification_report(y_test, y_pred))
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred))
print(f"AUC-ROC: {roc_auc_score(y_test, logreg.predict_proba(X_test)[:, 1]):.4f}")
```
